```python
import pandas as pd
# data manipulation-> read, slice/dice, rearrange data!
# return values-> DATAFRAME -> rectangular dataset -> rows and columns-> TABLE

data = pd.read_csv("https://raw.githubusercontent.com/a-forty-two/cylons/master/data.csv")
```

Column Selection

```python
cols_to_select = ['id', 'diagnosis']
newdata = data [ cols_to_select ]
newdata.head()
```

Row Selection

Know the column names

```python
# syntax = variable.loc [ row_selection, column_selection_by_column_names]
# starting from 0-> 3rd row, and 3 columns-> id, diagnosis and radius_mean
newdata = data.loc[ 2, ['id', 'diagnosis', 'radius_mean']]
newdata.head()
```

```python
# From 10th Row (11th from 0), till 12th Row 
# from, to 
newdata = data.loc[ 10:12, ['id', 'radius_mean']]
newdata.head()
```

```python
# From beginning, till 12th Row 
# from, to 
newdata = data.loc[ :12, ['id', 'radius_mean']]
newdata
```

```python
# From  560th Row , till the end
# from, to 
newdata = data.loc[ 560:, ['id', 'radius_mean']]
newdata
```

Know the column index

```python
# All rows represented by :
# only columns 2 to 4 [5 excluded]
newdata = data.iloc[:, 2:5]
newdata
```

```python
# if we use negative indexing, calculation is from bottom and right
# -5: will give me last 5 records (rows)
# -3 will give me last three columns 
newdata = data.iloc[-5:, -3:]
newdata
```
